# 1. Write Python scripts for basic file operations and data processing?

**Creating and Writing to a File**

```python
with open("example.txt", "w") as file:
    file.write("Hello, Python!\n")
    file.write("This is a sample file.")
```

**Reading a File**

```python
with open("example.txt", "r") as file:
    content = file.read()
    print(content)
```

**Appending Data to a File**

```python
with open("example.txt", "a") as file:
    file.write("\nAppending new data!")
```

**Processing Data from a CSV File**

```python
# Create a dummy data.csv file
with open("data.csv", "w") as file:
    file.write("name,age,city\n")
    file.write("Alice,30,New York\n")
    file.write("Bob,25,London\n")
    file.write("Charlie,35,Paris\n")
```

```python
import csv

with open("data.csv", "r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)  # Prints each row as a list

```

**Writing Data to a CSV File**

```python
import csv

data = [["Name", "Age"], ["Alice", 25], ["Bob", 30]]

with open("output.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(data)
```

**Processing JSON Data**

```python
import json

data = {"name": "Himanshu", "age": 21, "language": "Python"}

with open("data.json", "w") as file:
    json.dump(data, file, indent=4)
```

**Reading JSON Data**

```python
import json

with open("data.json", "r") as file:
    data = json.load(file)
    print(data)
```

# 2.Develop a simple web scraper to extract data from a website?

You can build a simple web scraper using Python's requests and BeautifulSoup libraries. Here's a basic example to extract titles from a webpage:

**Step 1: Install Required Libraries**

```python
pip install requests beautifulsoup4
```

**Step 2: Write the Web Scraper**

```python
import requests
from bs4 import BeautifulSoup

# Define the URL to scrape
url = "https://colab.research.google.com/"  # Replace with the actual website

# Send a request to fetch the webpage content
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, "html.parser")

# Extract all titles (assuming they are inside <h2> tags)
titles = soup.find_all("h2")

# Print extracted titles
for title in titles:
    print(title.text)
```
