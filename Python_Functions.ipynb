# 2.Explain the concept of function arguments and parameters in Python.

In Python, parameters and arguments are essential components of functions, but they serve different roles.

**1. Parameters vs. Arguments**

* **Parameters**: Variables defined in a function declaration that act as placeholders.

* **Arguments**: Actual values passed to a function when calling it.

** 2. Types of Function Arguments**

Python supports multiple types of arguments:

* **a) Positional Arguments**

    * Arguments must be passed in the correct order.

````python
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet("Himanshu", 21)  # Correct order
````

**b) Keyword Arguments**

* Arguments are passed using parameter names, allowing flexibility in order.


````python
def greet(name, age):
    print(f"Hello {name}, you are {age} years old.")

greet(age=21, name="Himanshu")  # Order doesn't matter
````

**c) Default Arguments**

* Parameters can have default values if no argument is provided.

````python
def greet(name, age=18):
    print(f"Hello {name}, you are {age} years old.")

greet("Himanshu")  # Uses default age (18)
````

**d) Variable-Length Arguments (*args and **kwargs)**

* Allows passing multiple arguments dynamically.

````python
def add_numbers(*args):
    return sum(args)

print(add_numbers(1, 2, 3, 4))  # Output: 10
````

````python
def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="Himanshu", age=21, language="Python")
````

# 5.  What are iterators in Python and how do they differ from iterables?

In Python, iterators and iterables are closely related but serve different purposes.

**1. Iterable vs. Iterator**

Key 1.png

**2. Example of an Iterable**

* An iterable is an object that can be looped over, like a list:

````python
numbers = [1, 2, 3]  # List is an iterable
for num in numbers:
    print(num)  # Iterates over the list
````

**3. Example of an Iterator**

* An iterator is an object that remembers where it is during iteration:

````python
numbers = [1, 2, 3]
iterator = iter(numbers)  # Convert list to an iterator

print(next(iterator))  # Output: 1
print(next(iterator))  # Output: 2
print(next(iterator))  # Output: 3
````

If next(iterator) is called again, it raises a StopIteration error.

**4. Creating a Custom Iterator**

* You can create a custom iterator by defining __iter__() and __next__():

````python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.current = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.current < self.limit:
            self.current += 1
            return self.current
        else:
            raise StopIteration

counter = Counter(3)
for num in counter:
    print(num)  # Output: 1, 2, 3
````

# 6.Explain the concept of generators in Python and how they are defined.

Generators in Python are a special type of function that allow lazy iteration, meaning they generate values on demand instead of storing them all in memory at once. This makes them highly efficient for handling large datasets or infinite sequences.

**How Generators Work**

* Instead of return, generators use yield, which pauses the function and retains its state.

* Each time the generator is called, execution resumes right after the last yield.

* Generators are memory-efficient because they don’t store all values at once.

**Defining a Generator Function**

* A generator function is defined like a normal function but uses yield instead of return:

````python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count  # Pauses execution and returns the value
        count += 1

# Using the generator
counter = count_up_to(5)
for num in counter:
    print(num)
````

**Key Differences Between yield and return**

Key 1.png

**Generator Expressions**

* Python also supports generator expressions, which are similar to list comprehensions but more memory-efficient:

````python
squares = (x*x for x in range(5))  # Generator expression
for num in squares:
    print(num)
````

# 7. What are the advantages of using generators over regular functions?

Generators offer several advantages over regular functions in Python, making them ideal for handling large datasets and optimizing memory usage. Here’s why they are beneficial:

**1. Memory Efficiency**

* Generators don’t store all values in memory; they generate values on demand using yield. This makes them perfect for working with large datasets or infinite sequences.

**2. Lazy Evaluation**

* Unlike regular functions that compute and return all values at once, generators pause execution at yield and resume when needed. This improves performance, especially in iterative processing.

**3. Faster Execution**

* Since generators don’t create large data structures, they execute faster than regular functions that return lists or other collections.

**4. Simplifies Code for Iteration**

* Generators eliminate the need for manual state tracking in loops. Instead of maintaining counters or indices, they remember their last execution point.

**5. Supports Infinite Sequences**

* Generators can produce infinite sequences without consuming excessive memory. For example, generating Fibonacci numbers:

````python
def fibonacci():
    a, b = 0, 1
    while True:
        yield a
        a, b = b, a + b

fib = fibonacci()
for _ in range(5):
    print(next(fib))  # Output: 0, 1, 1, 2, 3
````

**6. Improves Performance in Data Streaming**

* Generators are widely used in data streaming, where processing happens in chunks rather than loading everything at once.

# 8. What is a lambda function in Python and when is it typically used?

A lambda function in Python is a small, anonymous function that can have any number of arguments but only one expression. It is defined using the lambda keyword.

**Syntax of a Lambda Function**



```
lambda arguments: expression
```



**Example: Simple Lambda Function**

````python
square = lambda x: x ** 2
print(square(5))  # Output: 25
````

Here, lambda x: x ** 2 defines a function that squares a number.

**When to Use Lambda Functions?**

* Lambda functions are typically used when: ✅ You need a short, temporary function without defining a full function using def. ✅ You want to use a function inside another function (e.g., sorting, filtering). ✅ You need a function for functional programming (e.g., map(), filter(), reduce()).

````python
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
````

````python
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]
````

# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

The map(), reduce(), and filter() functions in Python are functional programming tools that help process iterables efficiently. Here’s how they differ:

**1. map() – Applies a Function to Each Element**

* ✅ Purpose: Transforms each element in an iterable using a function.
* ✅ Returns: A map object (iterator) containing transformed values.
* ✅ Example:

````python
numbers = [1, 2, 3, 4]
squared = list(map(lambda x: x ** 2, numbers))
print(squared)  # Output: [1, 4, 9, 16]
````

* ✅ Use Case: When you need to modify all elements in a list.

**2. filter() – Selects Elements Based on a Condition**

* ✅ Purpose: Filters elements based on a function that returns True or False.
* ✅ Returns: An iterator containing only the elements that satisfy the condition.
* ✅ Example:

````python
numbers = [1, 2, 3, 4, 5]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)  # Output: [2, 4]
````

* ✅ Use Case: When you need to remove unwanted elements from a list.

**3. reduce() – Applies a Function Cumulatively**

* ✅ Purpose: Reduces an iterable to a single value by applying a function cumulatively.
* ✅ Returns: A single computed value.
* ✅ Example:

````python
from functools import reduce

numbers = [1, 2, 3, 4]
product = reduce(lambda x, y: x * y, numbers)
print(product)  # Output: 24
````

* ✅ Use Case: When you need to aggregate values (e.g., sum, product, concatenation).

# 4.Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(n):
    if n < 2:
        return False  # Numbers less than 2 are not prime
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to sqrt(n)
        if n % i == 0:
            return False  # If divisible, it's not prime
    return True  # If no divisors found, it's prime

# Check prime numbers from 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
````

# 5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

````python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Counter to track iterations

    def __iter__(self):
        return self  # Returns the iterator object itself

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # Stops iteration when limit is reached
        value = self.a
        self.a, self.b = self.b, self.a + self.b  # Update Fibonacci numbers
        self.count += 1
        return value

# Example usage
fib_iterator = FibonacciIterator(10)
for num in fib_iterator:
    print(num)
````

# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
def power_of_two(exponent):
    result = 1
    for _ in range(exponent + 1):
        yield result  # Yield the current power of 2
        result *= 2  # Multiply by 2 for the next power

# Example usage
for power in power_of_two(5):
    print(power)
````

# 7. Implement a generator function that reads a file line by line and yields each line as a string.

````python
# Create a dummy file for demonstration
with open("example.txt", "w") as f:
    f.write("This is the first line.\n")
    f.write("This is the second line.\n")
    f.write("And here is the third line.\n")
````

````python
def read_file_line_by_line(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()  # Yield each line without newline characters

# Example usage
for line in read_file_line_by_line("example.txt"):
    print(line)
````

# 10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
def remove_vowels(string):
    vowels = {'a', 'e', 'i', 'o', 'u', 'A', 'E', 'I', 'O', 'U'}
    return ''.join(filter(lambda char: char not in vowels, string))

# Example usage
text = "Hello, Himanshu!"
print(remove_vowels(text))  # Output: "Hll, Hmns!"
````
