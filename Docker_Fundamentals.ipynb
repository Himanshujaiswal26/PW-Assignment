* **Now, First build the Docker-Compose.yaml file through below command**
```
docker-compose -f docker-compose.yaml build
```
6.png





* **Now, Down the Docker all the contain images through below command**
```
docker-compose -f docker-compose.yaml down
```
image.png



* **Now, UP the Docker all the contain images through below command and check the status**
```
docker-compose -f docker-compose.yaml up
```
