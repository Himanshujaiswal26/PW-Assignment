# 2. Explain the difference between mutable and immutable Data types with examples


In Python, mutable and immutable data types differ in whether their values can be changed after creation.

**Mutable Data Types**

* Can be modified after creation.

* Stored in memory at the same location even after modification.

* Examples: Lists, Dictionaries, Sets

**Example of a Mutable Data Type (List)**


```python
my_list = [1, 2, 3]
print("Before modification:", my_list)

my_list.append(4)  # Modifying the list
print("After modification:", my_list)  # Output: [1, 2, 3, 4]
```

**Immutable Data Types**


* Cannot be changed after creation.

* Any modification creates a new object in memory.

* Examples: Tuples, Strings, Integers, Floats

**Example of an Immutable Data Type (String)**

```python
text = "Hello"
print("Original:", text)

text = text + " World"  # A new string object is created
print("Modified:", text)  # Output: "Hello World"
```

# 8. What is a hash table, how does it relate to dictionaries in python?

A hash table is a data structure that stores key-value pairs using a hash function to map keys to specific locations in memory. This allows for fast lookups, insertions, and deletions, making hash tables highly efficient for data retrieval.

**How Hash Tables Work**

* The value associated with the key is stored at that index.

* When retrieving data, the hash function quickly finds the correct index, reducing search time.

# 17. Describe the time complexity of accessing elements in a dictionary.

Accessing elements in a Python dictionary is highly efficient due to its underlying hash table implementation.

**Time Complexity of Dictionary Access**

* **Average Case: O(1)** – Retrieving a value by its key is constant time because Python uses a hash function to map keys to memory locations.

* **Worst Case: O(n)** – In rare cases of hash collisions, multiple keys may be mapped to the same location, requiring a linear search within a bucket.

# 18. In what situations ar elists preferred over dictionaries?

Lists are preferred over dictionaries in Python when you need ordered, sequential data storage and index-based access. Here are some key situations where lists are the better choice:

**1. Maintaining Order**

# 19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Dictionaries in Python were traditionally considered unordered because they store key-value pairs using a hash table, which does not maintain a predictable sequence of elements. However, since Python 3.7, dictionaries preserve insertion order, meaning items are retrieved in the order they were added

**Why Were Dictionaries Unordered?**

* **Hashing Mechanism** – Keys are mapped to memory locations using a hash function, which does not inherently maintain order.

* **Efficient Lookups** – The focus is on fast access rather than sequential storage.

* **Dynamic Resizing** – Dictionaries adjust their internal structure as elements are added or removed, which can disrupt order.


**How Does This Affect Data Retrieval?**

* **Pre-Python 3.7** – Iterating over a dictionary could return items in an unpredictable order.

* **Python 3.7+** – Dictionaries maintain insertion order, making iteration more predictable.

* **Key-Based Access** – Since dictionaries use O(1) time complexity for lookups, order does not impact retrieval speed.
