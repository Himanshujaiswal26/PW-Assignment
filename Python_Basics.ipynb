# 2. Describe the role of predefined keywords in Python and provide examples of how they are used in a program

In Python, Predefined keywords are reserved words that have a special meaning in the language. They cannot be used as identifiers (variable names, function names, etc.). These keywords help define the syntax and structure of Python programs.

**Role of Predefined Keywords**

1. **Defining Control Flow –** Keywords like if, else, while, and for help control program execution.

2. **Creating Functions & Classes –** def and class are used to define functions and classes, respectively.

3. **Handling Exceptions –** try, except, and finally manage errors gracefully.

4. **Managing Imports & Modules –** import allows access to built-in and external modules.

5. **Working with Data & Variables –** global, nonlocal, and return modify variable scopes and return values.




**Example in python for Predefined Keywords usage**

```python
# Example using control flow keywords
def check_even(num):
    if num % 2 == 0:  # 'if' keyword controls decision-making
        return "Even"
    else:  # 'else' keyword defines an alternative path
        return "Odd"

print(check_even(10))  # Output: Even

# Example using loop and exception handling keywords
try:
    for i in range(5):  # 'for' keyword starts a loop
        print(i)
    raise ValueError("An error occurred")  # Manually raising an exception
except ValueError as e:
    print("Caught an error:", e)  # 'except' keyword handles errors
```

# 3.Complare and contrast mutable and immutable objects in Python with examples

In Python, objects are classified as mutable or immutable based on whether their state can be changed after creation.

**Mutable Objects**

*  Can be modified after they are created.

    *  **Examples:** Lists, dictionaries, sets, bytearrays

*  Changing a mutable object affects the original object itself.

**Examples of Mutable Objects(Lists)**

```python
numbers = [1, 2, 3]
print("Before modification:", numbers)

numbers.append(4)  # Modifying the list
print("After modification:", numbers)  # Output: [1, 2, 3, 4]
```

**Immutable Objects**


*  Cannot be changed after creation.

    *  Examples: Tuples, strings, integers, floats, booleans

*  When attempting to modify an immutable object, Python creates a new object instead.

**Example of an Immutable Object (String)**

```python
text = "Hello"
print("Original:", text)

text = text + " World"  # A new string object is created
print("Modified:", text)  # Output: "Hello World"
```

# 4. Discuss the different tyoes of operators in Python and provide examples of how they are used.


Python provides a variety of operators that allow developers to perform operations on variables and values. These operators can be categorized into different types:

**1. Arithmetic Operators**

Used for mathematical calculations like addition, subtraction, multiplication, etc.

```python
a = 10
b = 3
print(a + b)  # Addition: 13
print(a - b)  # Subtraction: 7
print(a * b)  # Multiplication: 30
print(a / b)  # Division: 3.33
print(a // b) # Floor Division: 3
print(a % b)  # Modulus: 1
print(a ** b) # Exponentiation: 1000
```

**2. Comparison Operators**

Used to compare values and return True or False.

```python
x = 5
y = 10
print(x > y)  # False
print(x < y)  # True
print(x == y) # False
print(x != y) # True
print(x >= y) # False
print(x <= y) # True
```

**3. Logical Operators**

Used to combine conditional statements.

```python
a = True
b = False
print(a and b)  # False
print(a or b)   # True
print(not a)    # False
```

**4. Bitwise Operators**

Operate at the binary level.

```python
a = 5  # 0101 in binary
b = 3  # 0011 in binary
print(a & b)  # Bitwise AND: 1
print(a | b)  # Bitwise OR: 7
print(a ^ b)  # Bitwise XOR: 6
print(~a)     # Bitwise NOT: -6
print(a << 1) # Left Shift: 10
print(a >> 1) # Right Shift: 2
```

**5. Assignment Operators**

Used to assign values to variables.

```python
x = 10
x += 5  # x = x + 5 → 15
x -= 3  # x = x - 3 → 12
x *= 2  # x = x * 2 → 24
x /= 4  # x = x / 4 → 6.0
```

**6. Identity Operators**

Check whether two variables refer to the same object.


```python
a = [1, 2, 3]
b = a
print(a is b)  # True
print(a is not b)  # False
```

**7. Membership Operators**

Check whether a value exists in a sequence.


```python
numbers = [1, 2, 3, 4, 5]
print(3 in numbers)  # True
print(6 not in numbers)  # True
```

# 5. Explain the concept of type casting in Python with examples.

Type casting in Python refers to **converting one data type into another.** It can be done in two ways:

**1. Implicit Type Casting (Automatic Conversion)**

Python automatically converts one data type to another when needed, without user intervention.

**Example: Integer to Float Conversion**


```python
a = 5   # Integer
b = 2.5 # Float
c = a + b  # Python converts 'a' to float automatically
print(c)  # Output: 7.5
print(type(c))  # Output: <class 'float'>
```

Python promotes smaller data types (like int) to larger ones (like float) to prevent data loss.

**2. Explicit Type Casting (Manual Conversion)**

Users manually convert data types using built-in functions like int(), float(), and str().


**Example: String to Integer Conversion**


```python
num_str = "10"  # String
num_int = int(num_str)  # Convert string to integer
print(num_int)  # Output: 10
print(type(num_int))  # Output: <class 'int'>
```

**Example: Float to Integer Conversion**


```python
num_float = 7.9
num_int = int(num_float)  # Converts float to integer (truncates decimal part)
print(num_int)  # Output: 7
```

**Key Takeaways**


* Implicit Casting happens automatically when Python determines it's safe.

* Explicit Casting requires manual intervention using functions like int(), float(), and str().

* Data Loss may occur in explicit casting (e.g., converting float to int removes decimal places).

# 6. How to conditional statements work in Python? Illustrate with Examples

Conditional statements in Python allow a program to make decisions based on specific conditions. They control the flow of execution by executing different blocks of code depending on whether a condition evaluates to **True or False.**

**Types of Conditional Statements in Python**

1. **if Statement** – Executes a block of code if the condition is true.

2. **if-else Statement** – Provides an alternative block of code if the condition is false.

3. **if-elif-else Statement** – Checks multiple conditions sequentially.

4. **Nested if Statements** – Places one if statement inside another.



**Examples**

**1. Simple if Statement**

```python
age = 20
if age >= 18:  # Condition is True
    print("Eligible to vote.")  # Output: Eligible to vote.
```

**2. if-else Statement**

```python
num = 10
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

**3. if-elif-else Statement**

```python
marks = 85
if marks >= 90:
    print("Grade: A")
elif marks >= 75:
    print("Grade: B")
else:
    print("Grade: C")
```

**4. Nested if Statement**

```python
x = 15
if x > 10:
    if x < 20:
        print("x is between 10 and 20")
```

# 7.Describe the different types of loops in Python and their use cases with examples.

Loops in Python allow you to execute a block of code multiple times, making tasks like iteration and automation much easier. Python has three main types of loops:

**1. for Loop**

Used for iterating over sequences like lists, tuples, strings, or ranges.

**Example: Iterating Over a List**



```python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
```

**Use Case**: When you need to loop through elements in a collection.

**2. while Loop**

Executes a block of code as long as a condition remains True.

**Example: Counting with a while Loop**

```python
count = 1
while count <= 5:
    print(count)
    count += 1
```

**Use Case:** When you need to repeat an action until a condition changes.

**3. Nested Loops**

Loops inside loops, useful for working with multi-dimensional data.

**Example: Printing a Pattern**


```python
for i in range(3):
    for j in range(3):
        print(f"({i}, {j})", end=" ")
    print()

```

**Use Case:** When working with matrices or nested data structures.

**Loop Control Statements**

Python provides special statements to control loops:

* **break:** Exits the loop immediately.

* **continue**: Skips the current iteration and moves to the next.

* **pass**: Acts as a placeholder.

**Example:** Using break and continue

```python
for num in range(10):
    if num == 5:
        break  # Stops the loop when num is 5
    if num % 2 == 0:
        continue  # Skips even numbers
    print(num)
```

**Use Case:** When you need to control loop execution dynamically.
